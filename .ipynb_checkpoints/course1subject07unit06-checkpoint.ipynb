## Selecting data from a pandas DataFrame

## iLoc operatore to select rows ranges and columns ranges

### df.iloc[rows, columns]

```python
df.iloc[:, 2:6] # 2 inclusive, 6 exclusive
```

### df.iloc[::2, 2:]

Here we are using the slicing notation ::2 for the rows which selects every second row
from the first to and up to the last. We are then using the slicing notation 2: for the
columns which selects every column from position 2 to the last column of the
DataFrame.

```python
df.iloc[::2] # Every second row
```

```python
df.iloc[::2, 2:] # Every second row from starting row, and every second column from the column position 2
```

## The loc operatore to select rows and columns by label
### df.loc[rows, columns]

```python
df.loc[:, "Height (m)" : "First ascent"] # All rows with columns from height (m) to First ascent, Both inclusive
```

```python
df.loc[:, "Height (m)" : "First ascent" :2]# All rows with columns from height (m) to First ascent, Both inclusive with 2 steps every other column
```

## Boolean selction of rows using the [] operator

### == equility, & (and), | (or) and ~ (not)

```python
df["Parent mountain"] == "Mount Everest" # Important
```

```python
df[df["Parent mountain"] == "Mount Everest"] # Records of mountains whose parents mountain is Mount Everest
```

```python
df["First ascent"] > 1955 # Important
```

```python
df[df["First ascent"] > 1955]
```

```python
# Combine the above two queries

# (df["Parent mountain"] == "Mount Everest") # Put within ()
# (df["First ascent"] > 1955) # Put wintin ()
# & combine it with operator
# Put within df[] as follow

df[(df["Parent mountain"] == "Mount Everest") & (df["First ascent"] > 1955)]
```

```python
# Mountains with Mount Everest as parent with a first ascent after 1955
df[(df["Parent mountain"] == "Mount Everest") & (df["First ascent"] > 1955)]
```

## Boolean selection of rows and columns using the .loc operator

```python
df.loc[(df["Parent mountain"] == "Mount Everest") & (df["First ascent"] > 1955)]
```

```python
# loc is good to select columns to return

df.loc[(df["Parent mountain"] == "Mount Everest") & (df["First ascent"] > 1955), "Height (m)":"Range"]
```

## Column criteria in loc selection

```python
col_criteria = [True, False, False, False, True, True, False]
```

```python
df["Height (m)"] > 8300
```

```python
df.loc[df["Height (m)"] > 8300, col_criteria]
```
